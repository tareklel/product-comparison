```python
import pandas as pd
```

```python
# inspect category and brands names manually to create dictionary files

ounass = pd.read_csv('output/ounass-2023-06-18.csv')
farfetch = pd.read_csv('output/farfetch-2023-06-18.csv')
```

```python
brands = pd.DataFrame({'brands':ounass.brand.unique(), 'ounass':ounass.brand.unique()})
```

```python
ff_brands = pd.DataFrame({'brands':farfetch.brand.unique(), 'farfetch':farfetch.brand.unique()})
ff_brands
```

```python
# left join on brands
brands = brands.merge(ff_brands, on='brands', how='left')
brands
```

```python
ff_brands['farfetch'][ff_brands['farfetch'].apply(lambda x: 'oose' in x)]
```

```python
brands.loc[brands.brands=='KENZO', 'farfetch'] = 'Kenzo'
brands.loc[brands.brands=='PUMA', 'farfetch'] = 'Puma'
brands.loc[brands.brands=='SAINT LAURENT', 'farfetch'] = 'Saint Laurent'
brands.loc[brands.brands=='Adidas', 'farfetch'] = 'adidas'
brands.loc[brands.brands=='Golden Goose Deluxe Brand', 'farfetch'] = 'Golden Goose'
```

```python
category = pd.DataFrame({'category':['shoes'], 'ounass':['shoes'], 'farfetch':['Shoes']})
category
```

```python
category.to_csv('dictionary/category.csv', index=False)
brands.to_csv('dictionary/brand.csv', index=False)
```
